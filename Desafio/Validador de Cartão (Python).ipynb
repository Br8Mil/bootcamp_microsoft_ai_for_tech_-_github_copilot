```python
import re

def validar_luhn(card_number):
    """Valida um nmero de carto de crdito usando o algoritmo de Luhn."""
    card_number = card_number.replace(" ", "")  # Remover espaos

    if not card_number.isdigit():
        return False  # Garantir que a entrada contm apenas dgitos

    total = 0
    reverse_digits = card_number[::-1]  # Inverter os dgitos

    # Algoritmo de Luhn: somar e verificar
    for i, digit in enumerate(reverse_digits):
        n = int(digit)
        if i % 2 == 1:
            n *= 2
            if n > 9:
                n -= 9  # Se for maior que 9, subtrai 9
        total += n

    return total % 10 == 0  # Retorna True se a soma for divisvel por 10

def identificar_tipo_cartao(card_number):
    """Identifica o tipo de carto com base no nmero e prefixo."""
    card_number = card_number.replace(" ", "")  # Remover espaos

    # Define um dicionrio com prefixos e tipos de cartes
    tipos_cartao = {
        "Mastercard": [("51", "55", 16)],
        "Visa": [("4", "", 16)],
        "American Express": [("34", "37", 15)],
        "Diners Club": [("300", "305", 14), ("36", "", 14)],
        "Discover": [("6011", "", 16), ("65", "", 16)],
        "Enroute": [("014", "", 16), ("149", "", 16)],
        "JCB": [("2131", "", 16), ("1800", "", 16), ("35", "", 16)],
        "Voyager": [("8699", "", 16)],
        "Hipercard": [("606282", "", 16)],
        "Aura": [("50", "", 16)],
    }

    # Verifica o prefixo e a quantidade de dgitos
    for tipo, prefixos in tipos_cartao.items():
        for prefixo, sufixo, length in prefixos:
            if card_number.startswith(prefixo) and (not sufixo or card_number.startswith(sufixo)) and len(card_number) == length:
                return tipo

    return "Tipo de carto desconhecido"  # Retorna se no encontrar um tipo correspondente

def validar_cartao(card_number):
    """Valida o nmero do carto e retorna o tipo ou erro."""
    # Primeiramente, valida o nmero do carto com Luhn
    if not validar_luhn(card_number):
        return "Nmero do carto invlido de acordo com o algoritmo de Luhn."

    # Identifica o tipo do carto
    tipo_cartao = identificar_tipo_cartao(card_number)
    return f"Este  um carto {tipo_cartao}."

# Solicitar input do usurio
cartao = input("Digite o nmero do carto de crdito: ").strip()
print(validar_cartao(cartao))
```
